```haskell
--
-- âˆ†Q using Rational
--
import DeltaQ

-- needed for visualisation within iHaskell
import Graphics.Rendering.Chart.Renderable

-- useful for rendering Rationals with 'just enough' precision
asFloat :: Rational -> Float
asFloat = fromRational
```

# Visualisations

There are options to plot
 - Single CDFs (normal and inverse)
   - with/without quantile lines
 - multiple CDFs (normal and inverse)

```haskell
e0, e1, e2 :: DQ
e0 = choice 0.0001 never $ uniform 0.5 1.5 .>>. uniform 0.5 3.5
e1 = choice 0.05 never $ e0 .>>. wait 0.5
e2 = e0 ./\. e1
```

```haskell
toRenderable $ plotCDF "e0" e0
toRenderable $ plotCDFs "e0, e1, e2" [("e0", e0), ("e1", e1), ("e2", e2)]
```

```haskell
toRenderable $ plotInverseCDF "e0 (inverse)" e0
toRenderable $ plotInverseCDFs "e0, e1, e2 (inverse)" [("e0", e0), ("e1", e1), ("e2", e2)]
```

```haskell
toRenderable $ plotCDFWithQuantiles "e0 - quantiles" [0.25, 0.5, 0.75, 0.9, 0.95] e0
toRenderable $ plotInverseCDFWithQuantiles "e0 (inverse) - quantiles" [0.5, 0.9, 0.99, 0.999] e0
```
